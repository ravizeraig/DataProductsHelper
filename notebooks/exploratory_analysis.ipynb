```python
import pandas as pd # Importe pandas novamente se ainda não o fez nesta sessão do Jupyter

if conn: # 'conn' é o objeto de conexão que você obteve na célula anterior
    try:
        # Dicionário para armazenar todos os DataFrames
        dataframes = {}

        # Lista de tabelas para carregar
        tables_to_load = [
            "products",
            "sales",
            "salesgoals",
            "salespeople",
            "commercialprojects",
            "mediamaps"
        ]

        for table_name in tables_to_load:
            print(f"\nCarregando dados da tabela '{table_name}'...")
            query = f"SELECT * FROM {table_name};"
            df = pd.read_sql_query(query, conn)
            dataframes[f"df_{table_name}"] = df
            print(f"DataFrame 'df_{table_name}' carregado. Primeiras 5 linhas:")
            print(df.head())
            print(f"Total de linhas em df_{table_name}: {len(df)}")
            print("-" * 50) # Linha separadora para facilitar a leitura

    except Exception as e:
        print(f"Erro ao carregar dados para o DataFrame: {e}")
    finally:
        # Garante que a conexão seja fechada no final
        close_db_connection(conn)
        print("\nConexão com o banco de dados fechada.")
else:
    print("Não foi possível carregar dados, pois a conexão com o banco de dados falhou.")
```

```python
# --- Investigação de valor_final_brl = 0 em df_sales ---
print("--- Investigando vendas com 'valor_final_brl' igual a 0 em df_sales ---")

# Filtra as linhas onde o valor_final_brl é 0
sales_zero_value = dataframes['df_sales'][dataframes['df_sales']['valor_final_brl'] == 0]

print(f"Número total de registros com 'valor_final_brl' igual a 0: {len(sales_zero_value)}")

if not sales_zero_value.empty:
    print("\nPrimeiras 10 linhas das vendas com valor 0:")
    print(sales_zero_value.head(10))

    # Analisar outras colunas relevantes para entender o contexto
    print("\nContagem de valores para 'regiao' nestas vendas de valor 0:")
    print(sales_zero_value['regiao'].value_counts())

    print("\nContagem de valores para 'vendedor_id' nestas vendas de valor 0 (top 10):")
    print(sales_zero_value['vendedor_id'].value_counts().head(10))

    # Você também pode querer ver a distribuição de 'quantidade_vendida' para estes casos
    print("\nEstatísticas descritivas para 'quantidade_vendida' nestas vendas de valor 0:")
    print(sales_zero_value['quantidade_vendida'].describe())

else:
    print("Não foram encontrados registros com 'valor_final_brl' igual a 0.")

print("-" * 70)

# --- Investigação de valor_proposto_brl = 0 em df_commercialprojects ---
print("\n--- Investigando projetos com 'valor_proposto_brl' igual a 0 em df_commercialprojects ---")

# Filtra as linhas onde o valor_proposto_brl é 0
projects_zero_value = dataframes['df_commercialprojects'][dataframes['df_commercialprojects']['valor_proposto_brl'] == 0]

print(f"Número total de registros com 'valor_proposto_brl' igual a 0: {len(projects_zero_value)}")

if not projects_zero_value.empty:
    print("\nPrimeiras 10 linhas dos projetos com valor proposto 0:")
    print(projects_zero_value.head(10))

    # Analisar a coluna 'status_negociacao' que é muito relevante aqui
    print("\nContagem de valores para 'status_negociacao' nestes projetos de valor 0:")
    print(projects_zero_value['status_negociacao'].value_counts())

    # Você pode também querer ver quais produtos_id estão associados a esses projetos de valor 0
    print("\nContagem de 'produto_id's nestes projetos de valor 0 (top 10):")
    print(projects_zero_value['produto_id'].value_counts().head(10))

else:
    print("Não foram encontrados registros com 'valor_proposto_brl' igual a 0.")

print("-" * 70)
```

```python
print("--- Verificando a ligação entre vendas de valor zero e projetos perdidos ---")

# 1. Filtrar df_sales para vendas com valor_final_brl = 0
sales_zero_value = dataframes['df_sales'][dataframes['df_sales']['valor_final_brl'] == 0]
print(f"Número de vendas com valor 0: {len(sales_zero_value)}")

# 2. Filtrar df_commercialprojects para projetos com valor_proposto_brl = 0 E status_negociacao = 'Perdido'
projects_lost_zero_value = dataframes['df_commercialprojects'][
    (dataframes['df_commercialprojects']['valor_proposto_brl'] == 0) &
    (dataframes['df_commercialprojects']['status_negociacao'] == 'Perdido')
]
print(f"Número de projetos perdidos com valor 0: {len(projects_lost_zero_value)}")

# 3. Realizar o merge (join) entre os dois DataFrames filtrados usando 'projeto_id'
# Usamos um 'inner' join para pegar apenas os registros que existem em AMBOS os lados
merged_zero_records = pd.merge(
    sales_zero_value,
    projects_lost_zero_value,
    left_on='projeto_id',
    right_on='id', # 'id' na tabela de projetos corresponde ao 'projeto_id' em vendas
    how='inner',
    suffixes=('_sales', '_projects') # Para diferenciar colunas com nomes iguais, como 'id'
)

print(f"\nNúmero de registros que coincidem (venda valor 0 e projeto perdido valor 0): {len(merged_zero_records)}")

if len(merged_zero_records) == len(sales_zero_value) and \
   len(merged_zero_records) == len(projects_lost_zero_value):
    print("CONFIRMAÇÃO: Todos os registros de vendas com valor 0 correspondem exatamente a projetos perdidos com valor 0.")
    print("Isso valida a lógica de que vendas de valor zero estão ligadas a projetos perdidos.")
else:
    print("ATENÇÃO: Não houve uma correspondência exata entre vendas de valor 0 e projetos perdidos com valor 0.")
    if not merged_zero_records.empty:
        print("\nAlguns registros que coincidem (primeiras 5 linhas):")
        print(merged_zero_records.head())

print("-" * 70)
```
