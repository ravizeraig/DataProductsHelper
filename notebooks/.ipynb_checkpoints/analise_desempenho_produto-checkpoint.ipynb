```python
# Célula 1: Importação de Bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px 
from ipywidgets import interact, widgets, Layout # Importar os widgets necessários
from IPython.display import display, clear_output # Para exibir e limpar saídas no notebook

```

```python
# Célula 2: Carregamento do DataFrame e Visão Geral - desempenho_pro_produto.csv

# --- Configurar o Caminho para os Dados Processados ---
processed_tables_dir = os.path.join('..', 'data', 'output', 'processed_tables')

# --- Carregar o DataFrame Específico: desempenho_pro_produto.csv ---
file_path = os.path.join(processed_tables_dir, 'desempenho_por_produto.csv')

# Verifica se o arquivo existe antes de tentar carregar
if not os.path.exists(file_path):
    print(f"ATENÇÃO: O arquivo '{os.path.basename(file_path)}' não foi encontrado em '{processed_tables_dir}'.")
    print("Por favor, certifique-se de que a tabela 'desempenho_por_produto.csv' foi gerada corretamente.")
    # Você pode adicionar um sys.exit() ou raise FileNotFoundError aqui se quiser parar a execução
else:
    df_desempenho_produto = pd.read_csv(file_path)

    print(f"DataFrame '{os.path.basename(file_path)}' carregado com sucesso!")
```

```python
print("\n--- Primeiras 5 linhas (df.head()) ---")
print(df_desempenho_produto.head())

print("\n--- Informações do DataFrame (df.info()) ---")
df_desempenho_produto.info()

print("\n--- Estatísticas Descritivas (df.describe()) ---")
print(df_desempenho_produto.describe())
```

```python
print("--- Primeiras 5 linhas do DataFrame de Desempenho de Produtos (df_desempenho_produto.head()) ---")
df_desempenho_produto.head()
```

```python
# Ordenar o DataFrame por quantidade vendida para que os subplots mostrem rankings
df_sorted_quantidade = df_desempenho_produto.sort_values(
    by='total_quantidade_vendida', ascending=False
).reset_index(drop=True)

num_produtos = len(df_sorted_quantidade)
produtos_por_subplot = 10

# Definindo a grade de subplots
ncols = 2  # Número de colunas de subplots (2 colunas)
nrows = (num_produtos // produtos_por_subplot) // ncols # 100 produtos / 10 por subplot = 10 subplots. 10 subplots / 2 colunas = 5 linhas.
# Ajustar altura do subplot. Talvez um pouco mais de altura por subplot ajude com rótulos de 90 graus.
fig, axes = plt.subplots(
    nrows=nrows,
    ncols=ncols,
    figsize=(16, 8 * nrows), # Largura 16 (8 por coluna), Altura 8 por linha (5 linhas = 40)
    sharey=False # Não compartilhar o eixo Y (nomes dos produtos)
)

# Ajustar o layout para evitar sobreposição de títulos e rótulos
fig.suptitle('Quantidade Total Vendida por Produto (Ranking Completo)', fontsize=20, y=1.02) # Título geral

# Função para formatar números para os rótulos (já definida, mas replicada para garantia)
def format_numbers_for_label(num):
    if num >= 1e6:
        return f'{num/1e6:.1f}M'
    elif num >= 1e3:
        return f'{num/1e3:.0f}K'
    return f'{num:.0f}'

# Aplanar o array de eixos para facilitar a iteração
axes_flat = axes.flatten()

# Iterar sobre os dados para criar cada subplot
for i in range(nrows * ncols): # Iterar pelo número total de subplots
    start_index = i * produtos_por_subplot
    end_index = min((i + 1) * produtos_por_subplot, num_produtos)
    
    # Se não houver mais produtos para este subplot, pula
    if start_index >= num_produtos:
        # Esconde o subplot vazio para não aparecer na visualização
        axes_flat[i].set_visible(False)
        continue

    # Slice do DataFrame para o subplot atual
    subset_df = df_sorted_quantidade.iloc[start_index:end_index].copy()
    
    # Seleciona o eixo para o subplot atual
    ax = axes_flat[i]

    # Criar o gráfico de barras vertical
    sns.barplot(
        x='nome_produto',
        y='total_quantidade_vendida',
        data=subset_df,
        ax=ax,
        palette='Blues_d',
        hue='nome_produto',
        legend=False
    )
    
    ax.set_title(f'Ranking: Produtos {start_index + 1} a {end_index}', fontsize=14)
    ax.set_xlabel('Nome do Produto', fontsize=12)
    ax.set_ylabel('Quantidade Vendida', fontsize=12)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    # --- Correção para o UserWarning e melhor visualização ---
    # Rotacionar os rótulos do eixo X para 90 graus e diminuir a fonte
    # Usamos setp diretamente nas ticklabels para evitar o warning e controlar a fonte
    plt.setp(ax.get_xticklabels(), rotation=90, ha='center', fontsize=8) # ha='center' para 90 graus é bom

    # Adicionar os rótulos de quantidade em cima de cada barra
    for container in ax.containers:
        ax.bar_label(container, fmt=lambda x: format_numbers_for_label(x), padding=3, fontsize=9) # Fonte menor para rótulos de barra

    # Ajustar o limite do eixo Y para dar espaço aos rótulos
    max_val = subset_df['total_quantidade_vendida'].max()
    ax.set_ylim(0, max_val * 1.20) # Aumenta o limite superior em 20% para mais espaço para o rótulo em 90 graus

plt.tight_layout(rect=[0, 0.03, 1, 1.01]) # Ajusta o layout para evitar sobreposição, deixando espaço para o suptitle
plt.show()

print("\nVisualização de todos os 100 produtos por quantidade vendida gerada em subplots verticais lado a lado com rótulos e nomes ajustados!")
```

```python
print("--- Iniciando cálculo do Valor Unitário Médio de Venda por Produto ---")

# --- Calcular o Valor Unitário Médio de Venda ---
# Criar uma cópia para evitar SettingWithCopyWarning
df_valor_unitario_venda = df_desempenho_produto.copy()

# Lidar com casos onde 'total_quantidade_vendida' pode ser 0 para evitar divisão por zero
# Substituir 0 por NaN ou uma pequena constante, ou simplesmente ignorar linhas com 0
# Para valor unitário, se a quantidade é 0, a venda não é relevante para preço unitário.
# Vamos filtrar essas linhas.
df_valor_unitario_venda = df_valor_unitario_venda[
    df_valor_unitario_venda['total_quantidade_vendida'] > 0
].copy()

df_valor_unitario_venda['valor_unitario_medio_brl'] = (
    df_valor_unitario_venda['total_valor_vendas_brl'] / 
    df_valor_unitario_venda['total_quantidade_vendida']
)

# Ordenar do mais caro (valor unitário médio de venda) para o mais barato
df_valor_unitario_venda = df_valor_unitario_venda.sort_values(
    by='valor_unitario_medio_brl', ascending=False
).reset_index(drop=True)

print("\n--- Primeiras linhas do DataFrame com Valor Unitário Médio de Venda ---")
print(df_valor_unitario_venda.head())
print(f"Total de produtos com valor unitário médio calculado: {len(df_valor_unitario_venda)}")
```

```python
# Usar o DataFrame recém-criado
df_to_plot = df_valor_unitario_venda.copy()

num_produtos = len(df_to_plot)
produtos_por_subplot = 10

ncols = 2
nrows = (num_produtos + produtos_por_subplot - 1) // produtos_por_subplot // ncols
if nrows == 0 and num_produtos > 0:
    nrows = 1 # Garante pelo menos 1 linha se houver produtos

fig, axes = plt.subplots(
    nrows=nrows,
    ncols=ncols,
    figsize=(16, 8 * nrows), # Mantendo o mesmo figsize para consistência
    sharey=False # Cada subplot terá sua própria escala de Y para os valores
)

fig.suptitle('Valor Unitário Médio de Venda por Produto (Ranking do Mais Caro ao Mais Barato)', fontsize=20, y=1.02)

# Função para formatar valores monetários para os rótulos
def format_currency_for_label(num):
    if num >= 1e6:
        return f'R${num/1e6:.1f}M'
    elif num >= 1e3:
        return f'R${num/1e3:.0f}K'
    return f'R${num:.0f}'

# Aplanar o array de eixos para facilitar a iteração
axes_flat = axes.flatten()

for i in range(nrows * ncols):
    start_index = i * produtos_por_subplot
    end_index = min((i + 1) * produtos_por_subplot, num_produtos)
    
    if start_index >= num_produtos:
        # Esconde o subplot vazio para não aparecer na visualização
        axes_flat[i].set_visible(False)
        continue

    subset_df = df_to_plot.iloc[start_index:end_index].copy()
    ax = axes_flat[i]

    sns.barplot(
        x='nome_produto',
        y='valor_unitario_medio_brl',
        data=subset_df,
        ax=ax,
        palette='magma', # Paleta para valores monetários de venda
        hue='nome_produto',
        legend=False
    )
    
    ax.set_title(f'Ranking: Produtos {start_index + 1} a {end_index}', fontsize=14)
    ax.set_xlabel('Nome do Produto', fontsize=12)
    ax.set_ylabel('Valor Unitário Médio de Venda (BRL)', fontsize=12)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Rotacionar os rótulos do eixo X e diminuir a fonte
    plt.setp(ax.get_xticklabels(), rotation=90, ha='center', fontsize=8)

    # Adicionar os rótulos de valor em cima de cada barra
    for container in ax.containers:
        ax.bar_label(container, fmt=lambda x: format_currency_for_label(x), padding=3, fontsize=9)

    # Ajustar o limite do eixo Y para dar espaço aos rótulos
    max_val = subset_df['valor_unitario_medio_brl'].max()
    ax.set_ylim(0, max_val * 1.25) # Aumenta a margem superior em 25% para os rótulos

plt.tight_layout(rect=[0, 0.03, 1, 1.01])
plt.show()

print("\nVisualização do valor unitário médio de venda de todos os produtos gerada com sucesso!")
```

```python
# Usar o DataFrame de desempenho para plotar a margem de lucro
df_to_plot = df_desempenho_produto.copy()

# Ordenar do maior (margem de lucro) para o menor
df_to_plot = df_to_plot.sort_values(
    by='margem_lucro_percentual', ascending=False
).reset_index(drop=True)

num_produtos = len(df_to_plot)
produtos_por_subplot = 10

ncols = 2
# Calcula o número de linhas necessário para os subplots
nrows = (num_produtos + produtos_por_subplot - 1) // produtos_por_subplot // ncols
if nrows == 0 and num_produtos > 0:
    nrows = 1 # Garante pelo menos 1 linha se houver produtos

fig, axes = plt.subplots(
    nrows=nrows,
    ncols=ncols,
    figsize=(16, 8 * nrows), # Mantendo o mesmo figsize para consistência
    sharey=False # Cada subplot terá sua própria escala de Y para as porcentagens
)

fig.suptitle('Margem de Lucro Percentual por Produto (Ranking da Maior para Menor Margem)', fontsize=20, y=1.02)

# Função para formatar porcentagens para os rótulos
def format_percentage_for_label(num):
    return f'{num:.1f}%'

# Aplanar o array de eixos para facilitar a iteração
axes_flat = axes.flatten()

for i in range(nrows * ncols):
    start_index = i * produtos_por_subplot
    end_index = min((i + 1) * produtos_por_subplot, num_produtos)
    
    if start_index >= num_produtos:
        # Esconde o subplot vazio para não aparecer na visualização
        axes_flat[i].set_visible(False)
        continue

    subset_df = df_to_plot.iloc[start_index:end_index].copy()
    ax = axes_flat[i]

    sns.barplot(
        x='nome_produto',
        y='margem_lucro_percentual',
        data=subset_df,
        ax=ax,
        palette='viridis', # Nova paleta para margem de lucro
        hue='nome_produto',
        legend=False
    )
    
    ax.set_title(f'Ranking: Produtos {start_index + 1} a {end_index}', fontsize=14)
    ax.set_xlabel('Nome do Produto', fontsize=12)
    ax.set_ylabel('Margem de Lucro Percentual (%)', fontsize=12)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Rotacionar os rótulos do eixo X e diminuir a fonte
    plt.setp(ax.get_xticklabels(), rotation=90, ha='center', fontsize=8)

    # Adicionar os rótulos de porcentagem em cima de cada barra
    for container in ax.containers:
        ax.bar_label(container, fmt=lambda x: format_percentage_for_label(x), padding=3, fontsize=9)

    # Ajustar o limite do eixo Y para dar espaço aos rótulos e garantir que 0% seja a base
    max_val = subset_df['margem_lucro_percentual'].max()
    ax.set_ylim(0, max_val * 1.25) # Aumenta a margem superior em 25% para os rótulos

plt.tight_layout(rect=[0, 0.03, 1, 1.01])
plt.show()

print("\nVisualização da margem de lucro percentual de todos os produtos gerada com sucesso!")
```

```python
# Gráfico Interativo Plotly com Pesquisa de Produto (via ipywidgets) - REVISADO

print("--- Preparando Gráfico Interativo Plotly com Pesquisa de Produto ---")

# --- Carregar o DataFrame df_desempenho_produto.csv ---
# Garantir que o DataFrame está disponível.
try:
    if 'df_desempenho_produto' not in locals() and 'df_desempenho_produto' not in globals():
        processed_tables_dir = os.path.join('..', 'data', 'output', 'processed_tables')
        file_path_desempenho = os.path.join(processed_tables_dir, 'desempenho_por_produto.csv')
        df_desempenho_produto = pd.read_csv(file_path_desempenho)
        print(f"DataFrame '{os.path.basename(file_path_desempenho)}' carregado com sucesso para o Plotly com pesquisa!")
except Exception as e:
    print(f"Erro ao carregar df_desempenho_produto: {e}. Certifique-se de que o arquivo existe e o caminho está correto.")
    df_desempenho_produto = pd.DataFrame()
    if df_desempenho_produto.empty:
        print("Não foi possível prosseguir com o gráfico Plotly sem o DataFrame de desempenho.")
        raise SystemExit("DataFrame df_desempenho_produto não disponível.")


# --- 1. Preparar o DataFrame Base para o Gráfico ---
df_plot_interactive = df_desempenho_produto.copy()

# Calcular o Valor Unitário Médio de Venda
df_plot_interactive = df_plot_interactive[
    df_plot_interactive['total_quantidade_vendida'] > 0
].copy()

df_plot_interactive['valor_unitario_medio_brl'] = (
    df_plot_interactive['total_valor_vendas_brl'] / 
    df_plot_interactive['total_quantidade_vendida']
)

# Ordenar o DataFrame pela Margem de Lucro Percentual (para consistência visual inicial)
df_plot_interactive = df_plot_interactive.sort_values(
    by='margem_lucro_percentual', ascending=False
).reset_index(drop=True)


# --- 2. Criar os Widgets de Interação ---
search_input = widgets.Text(
    description='Pesquisar Produto:',
    placeholder='Digite o nome do produto...',
    layout=Layout(width='50%')
)

all_products = ['Todos os Produtos'] + sorted(df_plot_interactive['nome_produto'].tolist())
product_dropdown = widgets.Dropdown(
    options=all_products,
    value='Todos os Produtos',
    description='Selecionar:',
    layout=Layout(width='50%')
)

graph_output = widgets.Output()


# --- 3. Função para Gerar/Atualizar o Gráfico ---
def update_graph(search_term="", selected_product="Todos os Produtos"):
    with graph_output:
        clear_output(wait=True) # Limpa a saída anterior para redesenhar o gráfico

        filtered_df = df_plot_interactive.copy()

        # Lógica de filtragem: prioriza a seleção do dropdown, depois a pesquisa por texto
        if selected_product != "Todos os Produtos":
            filtered_df = filtered_df[filtered_df['nome_produto'] == selected_product]
        elif search_term:
            filtered_df = filtered_df[
                filtered_df['nome_produto'].str.contains(search_term, case=False, na=False)
            ]
        
        if filtered_df.empty:
            display(go.Figure().add_annotation(text="Nenhum produto encontrado com os critérios de busca/seleção.",
                                              xref="paper", yref="paper",
                                              x=0.5, y=0.5, showarrow=False,
                                              font=dict(size=16)))
            return

        fig = make_subplots(
            rows=3, cols=1,
            shared_xaxes=True,
            vertical_spacing=0.08,
            subplot_titles=(
                'Total de Quantidade Vendida', 
                'Valor Unitário Médio de Venda (BRL)', 
                'Margem de Lucro Percentual (%)'
            )
        )

        custom_data_cols = ['total_quantidade_vendida', 'total_valor_vendas_brl', 
                            'valor_unitario_medio_brl', 'margem_lucro_percentual']
        
        # Quantidade Vendida
        fig.add_trace(
            go.Bar(
                x=filtered_df['nome_produto'],
                y=filtered_df['total_quantidade_vendida'],
                name='Quantidade Vendida',
                marker_color='#1f77b4',
                hovertemplate='<b>Produto:</b> %{x}<br>' +
                              '<b>Quantidade Vendida:</b> %{y:,.0f}<br>' +
                              '<b>Valor Total Vendas:</b> R$%{customdata[1]:,.2f}<br>' +
                              '<b>Valor Unitário Médio:</b> R$%{customdata[2]:,.2f}<br>' +
                              '<b>Margem de Lucro:</b> %{customdata[3]:.2f}%<extra></extra>',
                customdata=filtered_df[custom_data_cols].values
            ),
            row=1, col=1
        )

        # Valor Unitário Médio de Venda
        fig.add_trace(
            go.Bar(
                x=filtered_df['nome_produto'],
                y=filtered_df['valor_unitario_medio_brl'],
                name='Valor Unitário Médio',
                marker_color='#ff7f0e',
                hovertemplate='<b>Produto:</b> %{x}<br>' +
                              '<b>Quantidade Vendida:</b> %{customdata[0]:,.0f}<br>' +
                              '<b>Valor Total Vendas:</b> R$%{customdata[1]:,.2f}<br>' +
                              '<b>Valor Unitário Médio:</b> R$%{y:,.2f}<br>' +
                              '<b>Margem de Lucro:</b> %{customdata[3]:.2f}%<extra></extra>',
                customdata=filtered_df[custom_data_cols].values
            ),
            row=2, col=1
        )

        # Margem de Lucro Percentual
        fig.add_trace(
            go.Bar(
                x=filtered_df['nome_produto'],
                y=filtered_df['margem_lucro_percentual'],
                name='Margem de Lucro',
                marker_color='#2ca02c',
                hovertemplate='<b>Produto:</b> %{x}<br>' +
                              '<b>Quantidade Vendida:</b> %{customdata[0]:,.0f}<br>' +
                              '<b>Valor Total Vendas:</b> R$%{customdata[1]:,.2f}<br>' +
                              '<b>Valor Unitário Médio:</b> R$%{customdata[2]:,.2f}<br>' +
                              '<b>Margem de Lucro:</b> %{y:.2f}%<extra></extra>',
                customdata=filtered_df[custom_data_cols].values
            ),
            row=3, col=1
        )

        fig.update_layout(
            title_text='Desempenho Detalhado dos Produtos (Ordenado por Margem de Lucro)',
            height=1000,
            showlegend=False,
            hovermode='x unified',
            xaxis_tickangle=-45
        )

        fig.update_yaxes(title_text="Quantidade", row=1, col=1)
        fig.update_yaxes(title_text="Valor (BRL)", row=2, col=1)
        fig.update_yaxes(title_text="Margem (%)", row=3, col=1, ticksuffix="%")
        fig.update_xaxes(tickfont=dict(size=8), row=3, col=1)
        
        # --- AQUI ESTÁ A MUDANÇA: Usar display(fig) ---
        display(fig)

# --- 4. Conectar Widgets à Função de Atualização ---
def on_search_text_change(change):
    update_graph(search_term=change['new'], selected_product="Todos os Produtos")
    product_dropdown.value = "Todos os Produtos"

search_input.observe(on_search_text_change, names='value')

def on_dropdown_change(change):
    update_graph(search_term="", selected_product=change['new'])
    search_input.value = ""

product_dropdown.observe(on_dropdown_change, names='value')

display(search_input, product_dropdown, graph_output)

update_graph()

print("\nGráfico interativo com pesquisa e seleção gerado. Use os campos acima para filtrar!")
```
